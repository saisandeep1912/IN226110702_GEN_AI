# Tools

```python
def calculator_tool(input_text):
    return str(eval(input_text))

print("Tool Output:", calculator_tool("25 * 4"))
```

# Agents

```python
def simple_agent(query):
    if "calculate" in query or "*" in query:
        return calculator_tool(query.replace("calculate", ""))
    else:
        return llm.invoke(query)

print(simple_agent("calculate 10 * 5"))
print(simple_agent("Explain AI"))
```
